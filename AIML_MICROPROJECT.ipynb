```python
# Step 2: Import libraries
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
 #Step 3: Load dataset
data = pd.read_excel("dataset.csv.xlsx")
```

```python
# Step 4: Split input and output
X = data[['Maths','Biology','Creativity']].values
y = data['Career'].values
```

```python
# Step 5: One-hot encoding
y = to_categorical(y)
```

```python
# Step 6: Build ANN model
model = Sequential()
model.add(Dense(6, input_dim=3, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
# Step 7: Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Step 8: Train model
model.fit(X, y, epochs=100, verbose=0)
```

```python
# Step 9: Evaluate model
loss, accuracy = model.evaluate(X, y, verbose=0)
print("Model Accuracy:", accuracy)
```

```python

# Step 10: Take user input
print("\nEnter Student Details:")
maths = int(input("Maths Marks: "))
biology = int(input("Biology Marks: "))
creativity = int(input("Creativity Score: "))
```

```python
# Step 11: Prediction
sample = np.array([[maths, biology, creativity]])
prediction = model.predict(sample)
career = np.argmax(prediction)
```

```python
# Step 12: Output result
if career == 0:
    print("Recommended Career: Engineering")
elif career == 1:
    print("Recommended Career: Medical")
else:
    print("Recommended Career: Arts")
```
